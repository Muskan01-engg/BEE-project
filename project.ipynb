# **Que 8 :Which animated genre is produced the most?**

```python
from collections import Counter
def clean_and_count_genres(genres):
    return Counter(genre.strip() for genre in genres.split(','))
df['genre_counts'] = df['genres'].apply(clean_and_count_genres)
total_genre_counts = sum(df['genre_counts'], Counter())
for genre, count in total_genre_counts.items():
    print(f"{genre}: {count}")
```

With bargraph

```python
x = list(df['genre_counts'].sum().keys())[:11]
y = list(df['genre_counts'].sum().values())[:11]
plt.bar(x,y)
plt.title('Which genre had the most produced ?')
plt.xlabel('Genre')
plt.ylabel('Total Produced')
plt.xticks(rotation=90)
plt.show()
```
